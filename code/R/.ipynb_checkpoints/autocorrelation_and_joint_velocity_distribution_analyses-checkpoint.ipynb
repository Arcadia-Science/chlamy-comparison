## Prepare workspace and load data

<font size="3"> Set working directory.

```r
library(here)
here::here() # Should be set to the root of the repository
setwd(here::here())
```

<font size="3"> Source functions.

```r
source('code/R/chlamy_motility_utils.R')
```

<font size="3"> Load trajectory data for Experiment 1. These trajectories are the same as those used for motility analyses elsewhere in the pub and are available on Zenodo (zenodo.8326749). Here, we are analyzing full and unfiltered trajectories (as opposed to the focus filtered ones used for the comparisons of individual velocity parameters) which are provided as a .RDS file. 

```r
objs <- readRDS('code/R/data/chlamydomonas_species_full_motility_trajectories.RDS')
```

<font size="3"> Create data matrices for each species.

```r
# Combine all experiments into a single data frame
species <- do.call(rbind, objs)

# Remove NAs
species <- na.omit(species)

# Split on species
species <- split(as.data.frame(species), species$species)
```

<font size="3"> Split data into trials with single vs. multiple cells.

```r
# Multi
y <- duplicated(unlist(lapply(names(objs), function(x) substr(x, 1, nchar(x) - 2))))
for (i in 1:(length(y) - 1)) {
  if (y[i + 1] == TRUE) {
    y[i] <- TRUE
  }
}
multi <- objs[y == TRUE]

# Single
single <- objs[y == FALSE]

# Combine
multi <- do.call(rbind, multi)
single <- do.call(rbind, single)

# Remove NAs
multi <- na.omit(multi)
single <- na.omit(single)

# Split on species
single <- split(as.data.frame(single), single$species)
multi <- split(as.data.frame(multi), multi$species)
```

## Compare velocity autocorrelations

<font size="3"> Calculate and plot autocorrelations for single cell trials.

```r
# Initiate plot
options(repr.plot.width = 8, repr.plot.height = 8)

# Initiate plot with CR velocity autocorrelation
plot(acf(single$cr$velocity, lag = 100, plot = FALSE)$acf,
  type = "l",
  lwd = 2,
  ylim = c(-0.2, 1),
  ylab = "Autocorrelation",
  xlab = "Frames",
  col = "#73B5E3",
  lty = "dashed",
  cex.axis = 1.5,
  cex.lab = 1.5,
  bty = "n"
)

# Add CS velocity autocorrelation
lines(acf(single$cs$velocity, lag = 100, plot = FALSE)$acf,
  lwd = 2,
  col = "#97CD78",
  lty = "dashed"
)

# Add CR angular velocity autocorrelation
lines(acf(single$cr$angular_velocity, lag = 100, plot = FALSE)$acf,
  lwd = 2,
  col = "#73B5E3"
)

# Add CS angular velocity autocorrelation
lines(acf(single$cs$angular_velocity, lag = 100, plot = FALSE)$acf,
  lwd = 2,
  col = "#97CD78"
)

# Add line corresponding to autocorrelation of 0
abline(h = 0, lty = "dashed", col = "gray60")
```

<font size="3"> Calculate and plot autocorrelations for multiple cell trials.

```r
# Initiate plot
options(repr.plot.width = 8, repr.plot.height = 8)

# Initiate plot with CR velocity autocorrelation
plot(acf(multi$cr$velocity, lag = 100, plot = FALSE)$acf,
  type = "l",
  lwd = 2,
  ylim = c(-0.2, 1),
  ylab = "Autocorrelation",
  xlab = "Frames",
  col = "#73B5E3",
  lty = "dashed",
  cex.axis = 1.5,
  cex.lab = 1.5,
  bty = "n"
)

# Add CS velocity autocorrelation
lines(acf(multi$cs$velocity, lag = 100, plot = FALSE)$acf,
  lwd = 2,
  col = "#97CD78",
  lty = "dashed"
)

# Add CR angular velocity autocorrelation
lines(acf(multi$cr$angular_velocity, lag = 100, plot = FALSE)$acf,
  lwd = 2,
  col = "#73B5E3"
)

# Add CS angular velocity autocorrelation
lines(acf(multi$cs$angular_velocity, lag = 100, plot = FALSE)$acf,
  lwd = 2,
  col = "#97CD78"
)

# Add line corresponding to autocorrelation of 0
abline(h = 0, lty = "dashed", col = "gray60")
```

## Compare joint velocity distributions

<font size="3"> Extract single and multiple cell trials, this time not splitting data into trials (as was done above for the autocorrelation analyses).

```r
# Multi
y <- duplicated(unlist(lapply(names(objs), function(x) substr(x, 1, nchar(x) - 2))))
for (i in 1:(length(y) - 1)) {
  if (y[i + 1] == TRUE) {
    y[i] <- TRUE
  }
}
multi <- objs[y == TRUE]

# Single
single <- objs[y == FALSE]

# Combine
multi <- do.call(rbind, multi)
single <- do.call(rbind, single)

# Remove NAs
multi <- na.omit(multi)
single <- na.omit(single)
```

<font size="3"> Calculate joint velocity distributions (as probability density functions) for single cell experiments and plot.

```r
# Initiate plot
par(mfrow = c(1, 3))
options(repr.plot.width = 15, repr.plot.height = 5)

# Cr
cr1d <- plot_velocity_prob_density_function(single$velocity[single$species == "cr"],
  single$angular_velocity[single$species == "cr"],
  title_label = "Cr",
  pdf_n = 100
)

# Cs
cs1d <- plot_velocity_prob_density_function(single$velocity[single$species == "cs"],
  single$angular_velocity[single$species == "cs"],
  title_label = "Cs",
  pdf_n = 100
)

# Difference
d <- plot_diff_prob_density_function(cr1d,
  cs1d,
  col1 = "#97CD78",
  col2 = "#73B5E3"
)
```

<font size="3"> Calculate joint velocity distributions (as probability density functions) for multiple cell experiments and plot.

```r
# Initiate plot
par(mfrow = c(1, 3))
options(repr.plot.width = 15, repr.plot.height = 5)

# Cr
cr1d <- plot_velocity_prob_density_function(multi$velocity[multi$species == "cr"],
  multi$angular_velocity[multi$species == "cr"],
  title_label = "Cr",
  pdf_n = 100
)

# Cs
cs1d <- plot_velocity_prob_density_function(multi$velocity[multi$species == "cs"],
  multi$angular_velocity[multi$species == "cs"],
  title_label = "Cs",
  pdf_n = 100
)

# Difference
d <- plot_diff_prob_density_function(cr1d,
  cs1d,
  col1 = "#97CD78",
  col2 = "#73B5E3"
)
```
